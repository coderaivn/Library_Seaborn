```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cereal=pd.read_csv("/content/sample_data/cereal.csv")
df_cereal.head()
```

```python
df_cereal.shape
```

```python
df_cereal['mfr'].value_counts()
```

```python
list=["shelf","weight","cups","rating"]
df_cereal=df_cereal.drop(list,axis=1)
df_cereal.head()
```

```python
cereal_corr=df_cereal.corr()
cereal_corr
```

# Dng *heat map*
A heatmap is a plot of color-coded rectangular data. It helps show relationships between variables, including time.

```python
sns.heatmap(data=cereal_corr,cmap="Blues")
```

```python
# Make the data symmetric , one_like can build matrix of boolean (True , False) as  same data
matrix_one=np.ones_like(cereal_corr,dtype=bool)
matrix_one
```

```python
mark=np.triu(matrix_one)
mark
```

```python
adjusted_mark=mark[1:,:-1]
adjusted_mark
```

```python
fig ,ax = plt.subplots(figsize=(10,8))
#That method uses HUSL colors, so you need hue, saturation, and lightness.
#I used hsluv.org to select the colors of this chart.
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)
sns.heatmap(data=cereal_corr.iloc[1:, :-1],mask=adjusted_mark,annot=True,
            cmap=cmap,vmin=-1,vmax=1)
```
